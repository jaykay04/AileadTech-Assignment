## 2. Find the fastest flight and the most delayed flight.


```python
dataset
```

```python
#The fastest flight is the flight with the smallest arrival delay
dataset['arr_delay'].min()
```

```python
fastest_flight=dataset[dataset['arr_delay']==dataset['arr_delay'].min()]
fastest_flight
```

```python
#From the above, the fastest flight is the VX Carrier, flight 193 with an arrival delay time of -86
```

```python
#The most delayed flight is the flight that has the maximum departure delay
dataset['dep_delay'].max()
```

```python
most_delayed=dataset[dataset['dep_delay']==1301]
most_delayed
```

```python
#From here, the most delayed flight is the HA carrier, flight 51.
```

## Which flights traveled the longest? Which traveled the shortest?

```python
#The flight that travelled the longest is the flight with the highest distance covered
dataset['distance'].max()
```

```python
longest_travelled=dataset[dataset['distance']==dataset['distance'].max()]
longest_travelled
```

```python
longest_travelled['carrier'].unique()
```

```python
#The flight that travelled the longest is the HA carrier flight 51
```

```python
shortest_travelled=dataset[dataset['distance']==dataset['distance'].min()]
shortest_travelled
```

```python
#The flight that travelled the shortest is the US Carrier flight 1632.
```
